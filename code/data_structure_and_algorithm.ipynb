# 二分查找

```python
def binarySearch(arr, target):
    head = 0
    tail = len(arr) - 1
    while head <= tail:
        mid = (head + tail)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            head = mid + 1
        else:
            tail = mid - 1
    return "The target don't exist"

mylist = [1, 3, 5, 7, 9]
print(binarySearch(mylist, 7))
print(binarySearch(mylist, -1))
```

# 快速排序

```python
def quickSort(arr):
    if len(arr) < 2:
        return arr
    else:
        pivot = arr[0]
        less = [ele for ele in arr[1:] if ele <= pivot]
        greater = [ele for ele in arr[1:] if ele > pivot]
        return quickSort(less) + [pivot] + quickSort(greater)

print(quickSort([10, 5, 2, 1, 10, 34, 6, 9, 7, 3, 8]))
```

# 冒泡排序

```python
def bubbleSort(arr):
    for i in range(0, len(arr) - 1):
        for j in range(0, len(arr) - 1 - i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

if __name__ == '__main__':
    arr = [54, 26, 93, 77, 44, 31, 44, 55, 20]
    print('initial array: %s' % arr)
    bubbleSort(arr)
    print('Sorted array: %s' % arr)
```

# 选择排序

```python
# find the smallest element and it's corresponding index in the array 
def findsmallest(arr):
    smallest = arr[0]
    smallest_index = 0
    for i in range(len(arr)):
        if arr[i] < smallest:
            smallest = arr[i]
            smallest_index = i
    return smallest_index

# get the sorted array
def selectSort(arr):
    newarr = []
    for i in range(len(arr)):
        smallest_index = findsmallest(arr)
        newarr.append(arr.pop(smallest_index))
    return newarr

# test selection sort 
print(selectSort([5, 3, 6, 2, 10]))
print(selectSort([5, 6, 3, 2, 10, 12, 1, 3, 5, 8, 20]))
```

```python
# 方法一：直接使用字符串切片功能逆转字符串
def strReverse(strDemo):
    return strDemo[::-1] 
print(strReverse('pythontab.com'))

# 方法二：遍历构造列表法
def strReverse(strDemo): 
    strList=[] 
    for i in range(len(strDemo)-1, -1, -1): 
        strList.append(strDemo[i])
    return ''.join(strList)
print(strReverse('pythontab.com'))

# 方法三：使用reverse函数
def strReverse(strDemo): 
    strList = list(strDemo) 
    strList.reverse() 
    return ''.join(strList)
print(strReverse('pythontab.com'))

# 方法四：借助collections模块方法extendleft
import collections 
def strReverse(strDemo): 
    deque1=collections.deque(strDemo) 
    deque2=collections.deque() 
    for tmpChar in deque1: 
        deque2.extendleft(tmpChar) 
    return ''.join(deque2) 
print(strReverse('pythontab.com'))

# 方法五：递归实现
def strReverse(strDemo): 
    if len(strDemo)<=1: 
        return strDemo 
    return strDemo[-1]+strReverse(strDemo[:-1]) 
print(strReverse('pythontab.com'))

#方法六：借助基本的Swap操作，以中间为基准交换对称位置的字符
def strReverse(strDemo): 
    strList=list(strDemo) 
    if len(strList)==0 or len(strList)==1: 
        return strList 
    i=0 
    length=len(strList) 
    while i < length/2: 
        strList[i], strList[length-i-1]=strList[length-i-1], strList[i] 
        i+=1
    return ''.join(strList)
print(strReverse('pythontab.com'))
```

# Dijkstra算法
Dijkstra算法使用广度优先搜索算法解决赋权有向图或无向图的单源最短路径问题，算法最终得到一个最短路径树。该算法常用于路由算法或者作为其他图算法的一个子模块。
实现步骤：
1. 找出最便宜的节点，即在最短时间内到达的节点
2. 更新该节点对应的邻居节点的开销
3. 重复该过程，直到对图中的每个节点都这样做了
4. 计算最终路径

```python
# 方法一
import networkx as nx

def Dijkstra(G,start,end):
    RG = G.reverse(); dist = {}; previous = {}
    for v in RG.nodes():
        dist[v] = float('inf')
        previous[v] = 'none'
    dist[end] = 0
    u = end
    while u!=start:
        u = min(dist, key=dist.get)
        distu = dist[u]
        del dist[u]
        for u,v in RG.edges(u):
            if v in dist:
                alt = distu + RG[u][v]['weight']
                if alt < dist[v]:
                    dist[v] = alt
                    previous[v] = u
    path=(start,)
    last= start
    while last != end:
        nxt = previous[last]
        path += (nxt,)
        last = nxt
    return path

G=nx.DiGraph()
G.add_edge(0,1,weight=80)
G.add_edge(1,2,weight=50)
G.add_edge(1,3,weight=30)
G.add_edge(3,2,weight=10)
G.add_edge(2,4,weight=20)
G.add_edge(2,5,weight=30)
G.add_edge(4,5,weight=10)
G.add_edge(5,3,weight=5)
G.add_edge(2,6,weight=10)
G.add_edge(4,6,weight=10)
G.add_edge(3,6,weight=25)
G.add_edge(5,6,weight=35)
print(G.number_of_nodes())
print(G.number_of_edges())
print(G.nodes())
print(G.edges())
print(G.neighbors(1))

rs=Dijkstra(G,0,6)
print(rs)
```

# 广度优先搜索(BFS)与深度优先(DFS)

```python
#coding=utf-8 
class Gragh():
    def __init__(self,nodes,sides):
        '''
        nodes 表示点
        sides 表示边

        '''
        # self.sequense是字典，key是点，value是与key相连接的点
        self.sequense = {}
        # self.side是临时变量，主要用于保存与指定点相连接的点
        self.side = []
        for node in nodes:
            for side in sides:
                u, v = side
                # 指定点u与另一个点v在同一个边中，则说明这个点v与指定点u是相连接的点，则需要将这个点v放到self.side中
                if node == u:
                    self.side.append(v)
                elif node == v:
                    self.side.append(u)
            self.sequense[node] = self.side
            self.side=[]
        #print(self.sequense)


    '''
    # Depth-First-Search 
        深度优先算法，是一种用于遍历或搜索树或图的算法。沿着树的深度遍历树的节点，尽可能深的搜索树的分支。
        当节点v的所在边都己被探寻过，搜索将回溯到发现节点v的那条边的起始节点。
        这一过程一直进行到已发现从源节点可达的所有节点为止。如果还存在未被发现的节点，
        则选择其中一个作为源节点并重复以上过程，整个进程反复进行直到所有节点都被访问为止。属于盲目搜索。        
    '''
    def DFS(self,node0):
        #queue本质上是堆栈，用来存放需要进行遍历的数据
        #order里面存放的是具体的访问路径
        queue,order=[],[]
        #首先将初始遍历的节点放到queue中，表示将要从这个点开始遍历
        queue.append(node0)
        while queue:
            #从queue中pop出点v，然后从v点开始遍历了，所以可以将这个点pop出，然后将其放入order中
            #这里才是最有用的地方，pop（）表示弹出栈顶，由于下面的for循环不断的访问子节点，并将子节点压入堆栈，
            #也就保证了每次的栈顶弹出的顺序是下面的节点
            v = queue.pop()
            order.append(v)
            #这里开始遍历v的子节点
            for w in self.sequense[v]:
                #w既不属于queue也不属于order，意味着这个点没被访问过，所以将其放到queue中，然后后续进行访问
                if w not in order and w not in queue: 
                    queue.append(w)
        return order

    '''
     Breadth-First-Search
     BFS是从根节点开始，沿着树的宽度遍历树的节点。如果所有节点均被访问，则算法中止。
           广度优先搜索的实现一般采用open-closed表。
    '''
    def BFS(self,node0):
        #queue本质上是堆栈，用来存放需要进行遍历的数据
        #order里面存放的是具体的访问路径
        queue,order = [],[]
        #首先将初始遍历的节点放到queue中，表示将要从这个点开始遍历
        # 由于是广度优先，也就是先访问初始节点的所有的子节点，所以可以
        queue.append(node0)
        order.append(node0)
        while queue:
            #queue.pop(0)意味着是队列的方式出元素，就是先进先出，而下面的for循环将节点v的所有子节点
            #放到queue中，所以queue.pop(0)就实现了每次访问都是先将元素的子节点访问完毕，而不是优先叶子节点
            v = queue.pop(0)
            for w in self.sequense[v]:
                if w not in order:
                    # 这里可以直接order.append(w) 因为广度优先就是先访问节点的所有下级子节点，所以可以
                    # 将self.sequense[v]的值直接全部先给到order
                    order.append(w)
                    queue.append(w)
        return order




def main():  
    nodes = [i+1 for i in range(8)]

    sides=[(1, 2),
        (1, 3),
        (2, 4),
        (2, 5),
        (4, 8),
        (5, 8),
        (3, 6),
        (3, 7),
        (6, 7)]
    G = Gragh(nodes,sides)
    print(G.DFS(1))
    print(G.BFS(1))

if __name__ == "__main__":
    main() 
```

# 变位词判断

```python
# 方法一：排序比较法
def anagram(st1, st2):
    s1 = list(st1)
    s2 = list(st2)
    s1.sort()
    s2.sort()
    if s1 == s2:
        return True
    else:
        return False
    
print(anagram('abs', 'bsa'))

# 方法二：元素判断法
def anagram2(st1, st2):
    flag = 1
    for s in st1:
        if not s in list(st2):
            flag = 0
    if len(st1) != len(st2):
        flag = 0
    if flag == 1:
        return True
    else:
        return False
    
print(anagram2('abs', 'sba'))

# 方法三：计数比较法
def anagram3(st1, st2):
    nst1 = {}
    nst2 = {}
    for (s1, s2) in zip(st1, st2):
        nst1[s1] = st1.count(s1)
        nst2[s2] = st2.count(s2)
    if nst1 == nst2:
        return True
    else:
        return False
    
print(anagram('abs', 'sab'))
```

# 利用Python的list实现栈

```python
# python list实现栈
'''
栈的操作包括：
1. Stack()：创建一个新的空栈
2. stack.isEmpty()：判断是否为空
3. stack.push(item)：元素入栈
4. stack.peek()：元素出栈(不删除栈顶元素)
5. stack.size()：栈的大小
6. stack.pop()：元素出栈(删除栈顶元素)
'''

class Stack():
    def __init__(self):
        self.items = []
        
    def isEmpty(self):
        return self.items == []
    
    def push(self, item):
        self.items.append(item)
    
    def peek(self):
        if self.isEmpty():
            return "The stack is empty!"
        else:
            return self.items[-1]
    
    def size(self):
        return len(self.items)
    
    def pop(self):
        if self.isEmpty():
            return "The stack is empty!"
        else:        
            return self.items.pop()
    
# test
my_stack = Stack()
print(my_stack.isEmpty())
my_stack.push('h')
my_stack.push('a')
print(my_stack.size())
print(my_stack.isEmpty())
print(my_stack.peek())
print(my_stack.pop())
print(my_stack.peek())
print(my_stack.size())
print(my_stack.pop())
print(my_stack.size())
print(my_stack.isEmpty())
print(my_stack.peek())
print(my_stack.pop())
```

# 利用Python的list实现双端队列

```python
# Python list实现双端队
'''
双端队列操作
1. Deque()：创建双端队列
2. addFront(item)：在队首插入元素
3. addRear(item)：在队尾插入元素
4. rmFront()：从队首移除元素
5. rmRear()：从队尾移除元素
6. isEmpty()：判断双端队列是否为空
7. size()：返回双端队列的大小

'''
class Deque():
    def __init__(self):
        self.items = []
        
    def isEmpty(self):
        return self.items == []
    
    def addFront(self, item):
        self.items.append(item)
        
    def addRear(self, item):
        self.items.insert(0, item)
        
    def rmFront(self):
        if self.isEmpty():
            return "The deque is empty!"
        else:
            return self.items.pop()
        
    def rmRear(self):
        if self.isEmpty():
            return "The deque is empty!"
        else:
            return self.items.pop(0)
        
    def size(self):
        return len(self.items)
    
# test
d=Deque()

print(d.isEmpty())

d.addRear(4)

d.addRear('dog')

d.addFront('cat')

d.addFront(True)

print(d.size())

print(d.isEmpty())

d.addRear(8.4)

print(d.rmRear())

print(d.rmFront())
```

```python
from pythonds.basic.queue import Queue
 
def hotPotato(namelist, num):
    simqueue = Queue()
    for name in namelist:
        simqueue.enqueue(name)
 
    while simqueue.size() >1:
        for i in range(num):
            simqueue.enqueue(simqueue.dequeue())
 
        simqueue.dequeue()
 
    return simqueue.dequeue()
 
print(hotPotato(["Bill","David","Susan","Jane","Kent","Brad"],7))
```

```python
import random
from pythonds.basic.queue import Queue


class Printer:
    def __init__(self, ppm):
        self.pagerate = ppm
        self.currentTask = None
        self.timeRemaining = 0

    def tick(self):
        if self.currentTask != None:
            self.timeRemaining = self.timeRemaining - 1
            if self.timeRemaining <= 0:
                self.currentTask = None

    def busy(self):
        if self.currentTask != None:
            return True
        else:
            return False

    def startNext(self, newtask):
        self.currentTask = newtask
        self.timeRemaining = newtask.getPages() * 60 / self.pagerate


class Task:
    def __init__(self, time):
        # time为入队时间戳，pages为随机模拟的打印页数
        self.timestamp = time
        self.pages = random.randrange(1, 21)

    def getStamp(self):
        return self.timestamp

    def getPages(self):
        return self.pages

    def waitTime(self, currenttime):
        # currenttime为出队时间戳
        return currenttime - self.timestamp


def simulation(numSeconds, pagesPerMinute):
    labprinter = Printer(pagesPerMinute)
    printQueue = Queue()
    waitingtimes = []

    for currentSecond in range(numSeconds):
        if newPrintTask():
            task = Task(currentSecond)
            printQueue.enqueue(task)

        if (not labprinter.busy()) and (not printQueue.isEmpty()):
            nexttask = printQueue.dequeue()
            waitingtimes.append(nexttask.waitTime(currentSecond))
            labprinter.startNext(nexttask)

        labprinter.tick()

    averageWait = sum(waitingtimes) / len(waitingtimes)
    print("Average Wait %6.2f secs %3d tasks remaining." % (averageWait, printQueue.size()))


def newPrintTask():
    # 产生一个新的打印任务的概率
    num = random.randrange(1, 181)
    if num == 180:
        return True
    else:
        return False


for i in range(10):
    simulation(3600, 10)
```

# 递归转换整数到字符串

```python
# 递归转换整数到字符串
def to_str(n, base):
    convert_str = '0123456789ABCDEF'
    if n < base:
        return convert_str[n]
    else:
        return to_str(n//base, base) + convert_str[n%base]
print(to_str(10, 2))
```

# 判断回文字符串

```python
import string

def reverse_str(s):
    if len(s) == 1:
        return s
    else:
        return reverse_str(s[1:]) + s[0]

def isPalindrome(s):
    '''
    判断回文字符串需注意三点：
    1. 大小写
    2. 空格替换
    3. 标点符号替换
    其中标点符号替换可以使用string.punctuation,它是由在 C 语言区域中被视为
    标点符号的 ASCII 字符组成的字符串，其值为!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~
    '''
    s = s.lower()
    s = s.replace(' ', '')
    for cha in string.punctuation:
        s = s.replace(cha, '')
    return s == reverse_str(s)

print(isPalindrome('Reviled did I live, said I, as evil I did deliver'))
```

```python
import turtle

def drawTriangle(points,color,myTurtle):
    myTurtle.fillcolor(color)
    myTurtle.up()
    myTurtle.goto(points[0][0],points[0][1])
    myTurtle.down()
    myTurtle.begin_fill()
    myTurtle.goto(points[1][0],points[1][1])
    myTurtle.goto(points[2][0],points[2][1])
    myTurtle.goto(points[0][0],points[0][1])
    myTurtle.end_fill()

def getMid(p1,p2):
    return ( (p1[0]+p2[0]) / 2, (p1[1] + p2[1]) / 2)

def sierpinski(points,degree,myTurtle):
    colormap = ['blue','red','green','white','yellow',
                'violet','orange']
    drawTriangle(points,colormap[degree],myTurtle)
    if degree > 0:
        sierpinski([points[0],
                        getMid(points[0], points[1]),
                        getMid(points[0], points[2])],
                   degree-1, myTurtle)
        sierpinski([points[1],
                        getMid(points[0], points[1]),
                        getMid(points[1], points[2])],
                   degree-1, myTurtle)
        sierpinski([points[2],
                        getMid(points[2], points[1]),
                        getMid(points[0], points[2])],
                   degree-1, myTurtle)

def main():
    myTurtle = turtle.Turtle()
    myWin = turtle.Screen()
    myPoints = [[-100,-50],[0,100],[100,-50]]
    sierpinski(myPoints,3,myTurtle)
    myWin.exitonclick()

main()
```

# 顺序搜索

```python
# 有序列表的顺序搜索
def sequentialSearch(alist, item):
    pos = 0
    found = False
    while pos < len(alist) and not found:
        if alist[pos] == item:
            found = True
        else:
            pos += 1
    return found

testlist = [1, 2, 32, 8, 17, 19, 42, 13, 0]
print(sequentialSearch(testlist, 3))
print(sequentialSearch(testlist, 13))
```

```python
# 无序列表的顺序搜索
def orderedSequentialSearch(alist, item):
    pos = 0
    found = False
    stop = False
    while pos < len(alist) and not found and not stop:
        if alist[pos] == item:
            found = True
        elif alist[pos] > item:
            stop = True
        else:
            pos += 1
    return found

testlist = [0, 1, 2, 8, 13, 17, 19, 32, 42,]
print(orderedSequentialSearch(testlist, 3))
print(orderedSequentialSearch(testlist, 13))
```

```python
# 二分法搜索返回True或False
def binarySearch1(alist, item):
    head  = 0
    tail = len(alist) - 1
    found = False
    while head <= tail and not found:
        mid = (head + tail)//2
        if alist[mid] == item:
            found = True
        elif alist[mid] < item:
            head = mid + 1
        else:
            tail = mid - 1
    return found          

testlist = [0, 1, 2, 8, 13, 17, 19, 32, 42,]
print(binarySearch1(testlist, 3))
print(binarySearch1(testlist, 0))
print(binarySearch1(testlist, 1))
print(binarySearch1(testlist, 2))
print(binarySearch1(testlist, 8))
print(binarySearch1(testlist, 13))
print(binarySearch1(testlist, 17))
print(binarySearch1(testlist, 19))
print(binarySearch1(testlist, 32))
print(binarySearch1(testlist, 42))
```

# 散列表

```python
# 用ASCII数值散列一个字符串
def ahash(astring, tablesize):
    sum = 0
    for i in range(len(astring)):
        sum = sum + ord(astring[i])
    return sum%tablesize

print(ahash('cat', 11))
```

```python
# 用ASCII数值*权重散列一个字符串
def ahash2(astring, tablesize):
    sum = 0
    for i in range(len(astring)):
        sum = sum + ord(astring[i])*(i+1)
    return sum%tablesize

print(ahash2('cat', 11))
```

# 利用两个列表实现散列表类

```python
# 创建哈希表类
class Hashtable():
    def __init__(self):
        self.size = 11
        self.slots = [None]*self.size
        self.data = [None]*self.size
        
    # 定义put()方法
    def put(self, key, data):
        hashvalue = self.hashFunc(key, len(self.slots)) # hashvalue为计算出的一个整数，此处相当于列表的下标
        if  self.slots[hashvalue] == None: # 创建新的键值对
            self.slots[hashvalue] = key
            self.data[hashvalue] = data
        else:
            if self.slots[hashvalue] == key: # 该键的位置没有被rehash过，重新赋予键新值
                print('***', hashvalue, self.slots[hashvalue], key)
                self.data[hashvalue] = data
            else: # 该键的位置被rehash过，重新赋予键新值
                nextslot = self.rehash(hashvalue, len(self.slots))
                while self.slots[nextslot] != None and self.slots[nextslot] != key:
                    nextslot = self.rehash(nextslot, len(self.slots))
                if self.slots[nextslot] == None:
                    self.slots[nextslot] = key
                    self.data[nextslot] = data
                else:
                    self.data[nextslot] = data

    # 定义hashFunc()方法
    def hashFunc(self, key, size):
        return key%size

    # 定义rehash()方法
    def rehash(self, oldvalue, size):
        return (oldvalue + 1)%size

    # 定义get()方法
    def get(self, key):
        initslot = self.hashFunc(key, len(self.slots))
        data = None
        stop = False
        found = False
        pos = initslot
        while self.slots[pos] != None and not found and not stop:
            if self.slots[pos] == key:
                found = True
                data = self.data[pos]
            else:
                pos = self.rehash(pos, len(self.slots))
                if pos == initslot: # 循环到开始处，表示没有搜索到
                    stop = True
        return data

    def __getitem__(self, key):
        return self.get(key)

    def __setitem__(self, key, data):
        self.put(key, data)
        
# test
H=Hashtable()
H[54]="cat"      # 54%11 = 10
H[26]="dog"      # 26%11 = 4
H[93]="lion"     # 93%11 = 5
H[17]="tiger"    # 17%11 = 6
H[77]="bird"     # 77%11 = 0
H[31]="cow"      # 31%11 = 9
H[44]="goat"     # 44%11 = 0
H[55]="pig"      # 55%11 = 0
H[20]="chicken"  # 20%11 = 9
print(H.slots)   # [77, 44, 55, 20, 26, 93, 17, None, None, 31, 54]
print(H.data)    # ['bird', 'goat', 'pig', 'chicken', 'dog', 'lion', 'tiger', None, None, 'cow', 'cat']

print(H[20])     # 'chicken'
print(H[17])     # 'tiger'
H[20]="duck"
print(H[20])     # 'duck'
print(H.slots)
print(H.data)    # ['bird', 'goat', 'pig', 'duck', 'dog', 'lion','tiger', None, None, 'cow', 'cat']
H[26] = "awu!"
print(H.slots)
print(H.data)    # ['bird', 'goat', 'pig', 'duck', 'dog', 'lion','tiger', None, None, 'cow', 'cat']
H[26] = 'haha'
print(H.slots)
print(H.data)    # ['bird', 'goat', 'pig', 'duck', 'dog', 'lion','tiger', None, None, 'cow', 'cat']
print(H[99])     # None
```

# 冒泡排序

```python
# 冒泡排序
def bubbleSort(arr):
    for i in range(0, len(arr) - 1):
        for j in range(0, len(arr) - 1 - i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                
alist = [54,26,93,17,77,31,44,55,20]
bubbleSort(alist)
print(alist)
```

```python
# 短路冒泡排序
def shortBubbleSort(alist):
    '''
    exchange用来控制是否继续遍历比较，当发现alsit[i] > alist[i+1]
    一直成立时，说明列表已经是有序的了，因此不再进行遍历，此时exchange
    为False，所以停止遍历，整个排序完成
    '''
    exchange = True
    n = len(alist) - 1
    while n > 0 and exchange:
        exchange = False
        for i in range(n):
            if alist[i] > alist[i+1]:
                exchange = True
                alist[i], alist[i+1] = alist[i+1], alist[i]
        n -= 1
        
alist=[20,30,40,90,50,60,70,80,100,110]
shortBubbleSort(alist)
print(alist)
```

# 选择排序

```python
# 方法一：选择最小的数从左往右排序
def selectSort1(alist):
    n = len(alist)
    for i in range(n-1):
        pos = i
        for j in range(i+1, n):
            if alist[pos] > alist[j]:
                pos = j
        alist[pos], alist[i] = alist[i], alist[pos]
        
alist = [54,26,93,17,77,31,44,55,20]
selectSort1(alist)
print(alist)
```

```python
# 方法二：选择最大的数从右往左排序
def selectSort2(alist):
    for i in range(len(alist)-1, 0, -1):
        pos = 0
        for j in range(1, i+1):
            if alist[j] > alist[pos]:
                pos = j
        alist[pos], alist[i] = alist[i], alist[pos]

alist = [54,26,93,17,77,31,44,55,20]
selectSort2(alist)
print(alist)
```

# 插入排序

```python
# 插入排序
def insertSort(alist):
    for i in range(1, len(alist)):
        pos = i
        currentvalue = alist[i]
        while pos > 0 and currentvalue < alist[pos-1]:
            alist[pos] = alist[pos-1]
            pos -= 1
        alist[pos] =  currentvalue
        
alist = [54,26,93,17,77,31,44,55,20]
insertSort(alist)
print(alist)
```

# 希尔排序

```python
# 希尔排序
def shellSort(alist):
    sublistcount = len(alist)//2
    while sublistcount > 0:
        for startposition in range(sublistcount):
            gapInsertionSort(alist,startposition,sublistcount)
        print("After increments of size", sublistcount, ', ', "The list is", alist)
        sublistcount = sublistcount // 2

def gapInsertionSort(alist,start,gap):
    for i in range(start+gap,len(alist),gap):
        currentvalue = alist[i]
        position = i
        while position>=gap and alist[position-gap]>currentvalue:
            alist[position]=alist[position-gap]
            position = position-gap
        alist[position]=currentvalue

alist = [54,26,93,17,77,31,44,55,20]
shellSort(alist)
print(alist)
```

# 快速排序

```python
# 快速排序方法一
def quickSort1(alist):
    quickSortHelper(alist,0,len(alist)-1)

def quickSortHelper(alist,first,last):
    if first<last:
        splitpoint = partition(alist,first,last)
        quickSortHelper(alist,first,splitpoint-1)
        quickSortHelper(alist,splitpoint+1,last)

def partition(alist,first,last):
    pivotvalue = alist[first]

    leftmark = first+1
    rightmark = last

    done = False
    while not done:
        while leftmark <= rightmark and alist[leftmark] <= pivotvalue:
            leftmark = leftmark + 1

        while alist[rightmark] >= pivotvalue and rightmark >= leftmark:
            rightmark = rightmark -1

        if rightmark < leftmark:
            done = True
        else:
            temp = alist[leftmark]
            alist[leftmark] = alist[rightmark]
            alist[rightmark] = temp

    temp = alist[first]
    alist[first] = alist[rightmark]
    alist[rightmark] = temp
    return rightmark

alist = [54,26,93,17,77,31,44,55,20]
quickSort1(alist)
print(alist)
```

```python
# 快速排序方法二
def quickSort2(arr):
    if len(arr) < 2:
        return arr
    else:
        pivot = arr[0]
        less = [ele for ele in arr[1:] if ele <= pivot]
        greater = [ele for ele in arr[1:] if ele > pivot]
        return quickSort2(less) + [pivot] + quickSort2(greater)

alist = [54,26,93,17,77,31,44,55,20]
quickSort2(alist)
print(alist)
print(quickSort2(alist))
```

```python
# 快速排序方法三
def quickSort3(arr, left, right):
    if left >= right:
        return
    low = left
    high = right
    pivot = arr[low]
    while low < high:
        while low < high and arr[high] >= pivot:
            high -= 1
        arr[low] = arr[high]
        while low < high and arr[low] < pivot:
            low += 1
        arr[high] = arr[low]
    arr[low] = pivot
    quickSort3(arr, left, low-1)
    quickSort3(arr, low+1, right)

alist = [54,26,93,17,77,31,44,55,20]
quickSort3(alist, 0, len(alist)-1)
print(alist)
```

```python
# 快速排序方法四
def quickSort4(array, l, r):
    if l < r:
        q = partition(array, l, r)
        quickSort4(array, l, q - 1)
        quickSort4(array, q + 1, r)

def partition(array, l, r):
    x = array[r]
    i = l - 1
    for j in range(l, r):
        if array[j] <= x:
            i += 1
            array[i], array[j] = array[j], array[i]
    array[i + 1], array[r] = array[r], array[i+1]
    return i + 1

alist = [54,26,93,17,77,31,44,55,20]
quickSort4(alist, 0, len(alist)-1)
print(alist)
```

```python
# -*- coding: utf-8 -*-
# 定义结点类
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# 定义链表类
class SinglyLinkedList:
    def __init__(self):
        self.head  = None

     # 在链表头部插入结点
    def insertToHead(self, value):
        newNode = Node(value)
        newNode.next = self.head
        self.head = newNode

    # 在某一节点后插入结点
    def insertAfter(self, node, value):
        if node == None:
            return
        newNode = Node(value)
        newNode.next = node.next
        node.next = newNode

    # 在某一结点前插入结点
    def insertBefore(self, node, value):
        if node == None or self.head == None:
            return
        if node == self.head:
            self.insertToHead(value)
            return
        newNode = Node(value)
        current = self.head
        while current.next != node and current.next:
            current = current.next
        if not current.next:
            return
        else:
            current.next = newNode
            newNode.next = node

    # 删除某个结点
    def deleteNode(self, node):
        if not self.head: # 链表为空
            return "The LinkedList is empty!"
        if node == self.head: # 要删除的结点在链表头，满足链表只有一个结点的情况
            self.head = node.next
            return
        current = self.head
        while current.next and current.next != node:
            current = current.next
        if not current.next:
            return
        else:
            current.next = node.next

    # 按值删除结点
    def deleteVal(self, value):
        if self.head == None:
            return
        if self.head.data == value:
            self.head = self.head.next
            return
        pre = self.head
        current = self.head.next
        while current and current.data != value:
            pre = current
            current = current.next
        if not current:
            return
        else:
            pre.next = current.next
            
    def println(self):
        nums = []
        current = self.head
        while current:
            nums.append(current.data)
            current = current.next
        print("->".join([str(num) for num in nums]))
            
# test
# insertToHead
print('-'*10, 'insertToHead', '-'*10)
linklist = SinglyLinkedList()
for i in range(15):
    linklist.insertToHead(i)
linklist.println()

# insertAfter
print('-'*10, 'insertAfter', '-'*10)
linklist.println()
linklist.insertAfter(linklist.head, 100)
linklist.println()
linklist.insertAfter(linklist.head.next, 200)
linklist.println()
linklist.insertAfter(linklist.head.next.next, 300)
linklist.println()

# insertBefore
print('-'*10, 'insertBefore', '-'*10)
linklist.println()
linklist.insertBefore(linklist.head, 55)
linklist.println()
linklist.insertBefore(linklist.head.next, 555)
linklist.println()
linklist.insertBefore(linklist.head.next.next, 5555)
linklist.println()

# deleteNode
print('-'*10, 'deleteNode', '-'*10)
linklist.println()
linklist.deleteNode(linklist.head.next.next)
linklist.println()
linklist.deleteNode(linklist.head.next)
linklist.println()
linklist.deleteNode(linklist.head)
linklist.println()

# deleteVal
print('-'*10, 'deleteVal', '-'*10)
linklist.println()
linklist.deleteVal(100)
linklist.println()
linklist.deleteVal(200)
linklist.println()
linklist.deleteVal(300)
linklist.println()
```

# 链表的应用操作

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# reverse singly-linked list
# Note: the input is Node rather than a linked list
def reverseNode(head):
    if not head and not head.next:
        return head
    pre = None
    current = head
    while current:
        current.next, current, pre = pre, current.next, current
    return pre

# cycle detection in a list
def detectCycle(head): # 使用快慢指针法判断
    slow, fast = head, head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

# entry detection in a list
def detectEntry(head): # 使用快慢指针法判断
    '''
    Note:
        1. 当链表中有环时，返回环的入口以及环中的环的入口的上一个结点
        2. 当链表中无环时，返回None
    '''
    if not head and not head.next:
        return None
    slow, fast = head, head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break
    if fast == slow:
        fast = head
        while fast != slow:
            pre = slow
            fast = fast.next
            slow =  slow.next
        return fast, pre
    return None

# merger two sorted linked list
def mergeList(lst1, lst2):
    if lst1 and lst2:
        p1, p2 = lst1, lst2
        
        fakeHead = Node(None) # fakeHead为哨兵
        current = fakeHead
        while p1 and p2:
            if p1.data <= p2.data: # if判断两个链表中的较小的值，并用current.next指向它
                current.next = p1
                p1 = p1.next
            else:
                current.next = p2
                p2 = p2.next
            current = current.next
        current.next = p1 if p1 else p2 # 其中一个链表被遍历完
        return fakeHead.next
    return lst1 or lst2

# delete the nth node from the end whre n > 0
def deleteNode(head, n):
    first = head
    count = 0
    while first and count < n:
        first = first.next
        count += 1
    if not first and count  < n: # len(linkedList) < n
        return head
    if not first and count == n: # len(linkedList) = n
        return head.next

    # len(linkedList)  > n 
    second = head
    while first.next:
        first, second =  first.next, second.next
    second.next = second.next.next
    return head

# find middle node
def findMidNode(head):
    '''
    Note: 
        1. 所用方法：快慢指针法
        2. 分两种情况：链表长度为奇数或偶数
        3. 当链表长度为奇数时，快慢指针法的 while 循环结束条件是first.next -> None，而fast指向最后一个结点，其值不为空，
           此时slow正好指向中间结点，例如:
           --->[]--->[]--->[]--->[]--->[]---None
                      slow        first
        4. 当链表长度为偶数时，快慢指针法的 while 循环结束条件是first -> None，fast指向None，其值为空，
           此时slow指向两个中间结点的后一个结点，例如:
           --->[]--->[]--->[]--->[]---None
                          slow       first
        5. 当链表长度为偶数时，这时中间节点有两个，若要返回二者之中的前一个，
           请将带'***'代码注释掉，反则保持注释
    '''
    first, second = head, head
    #first = first.next if first else None  # (***)
    while first and first.next:
        first, second = first.next.next, second.next
    return second

# print the structure of linked list
def println(head):
    nums = []
    current = head
    pos = detectEntry(current)
    if not pos:
        while current:
            nums.append(current.data)
            current = current.next
        print("->".join([str(num) for num in nums]))
    else:
        pos[1].next = None # 这一步相当于解除环
        while current:
            nums.append(current.data)
            current = current.next
        print("->".join([str(num) for num in nums]))
    
    
# test
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)
node1.next = node2
node2.next = node3
node3.next = node4

# reverse
print('-'*10, 'reverse', '-'*10)
println(node1)
p = reverseNode(node1)
println(p)
p = reverseNode(p)
println(p)

# detectCycle & detectEntry
print('-'*10, 'detectCycle & detectEntry', '-'*10)
print(detectCycle(p))
pos = detectEntry(p)
if pos:
    print(pos[0].data, pos[1].data)
else:
    print(pos)
node5 = Node(5)
node4.next = node5
node5.next = node3
print(detectCycle(p))
pos = detectEntry(p)
print(pos[0].data, pos[1].data)
println(p)
print(detectCycle(p))

# mergeList
print('-'*10, 'mergeList', '-'*10)
lst1 = Node(0)
lst2 = Node(6)
lst1.next = lst2
node5.next = None
print(detectCycle(p))
println(p)
println(lst1)
longList = mergeList(lst1, p)
println(lst1)
println(longList)

# findMidNode
print('-'*10, 'findMidNode', '-'*10)
println(longList)
println(findMidNode(longList)) # 奇数个结点
_s = deleteNode(longList, 1)
println(longList)
println(findMidNode(longList)) # 偶数个结点
```

# 判断链表是否是回文链表

```python
# 判断链表是否是回文链表
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def isPalindrome(head: Node) -> bool:
    '''
    Note:
    1. 所用方法：快慢指针法
    2. 分两种情况：链表长度为奇数或偶数
    3. 当链表长度为奇数时，快慢指针法的 while 循环结束条件是first.next -> None，而fast指向最后一个结点，其值不为空，
       此时slow正好指向中间结点，例如:
       --->[]--->[]--->[]--->[]--->[]---None
                      slow        first
    4. 当链表长度为偶数时，快慢指针法的 while 循环结束条件是first -> None，fast指向None，其值为空，
       此时slow指向两个中间结点的后一个结点，例如:
       --->[]--->[]--->[]--->[]---None
                      slow       first
    '''
    if not head and not head.next:
        return False
    fast, slow = head, head
    pre = None
    while fast and fast.next:
        fast = fast.next.next
        slow.next, slow, pre = pre, slow.next, slow # 这一行有两个作用，一是慢指针移动，二是反转链表
        
    s1 = pre
    if fast: # 链表长度为奇数
        s2 = slow.next
    else: # 链表长度为偶数
        s2 = slow
    while s1 and s2:
        if s1.data != s2.data:
            return False
        s1 = s1.next
        s2 = s2.next
    return not s1 and not s2
    
node1 = Node('a')
node2 = Node('b')
node3 = Node('b')
node4 = Node('a')
node1.next = node2
node2.next = node3
node3.next = node4
print(isPalindrome(node1))
```

# 二分查找

```python
'''
二分查找的应用场景：
    1. 二分查找依赖的是数组
    2. 二分查找针对的是有序数据
    3. 针对插入、删除操作不频繁，一次排序多次查找的场景
    4. 数据量特别大不适用，因为其依赖于数组，对连续空间要求高

二分查找易错点
    1. 循环条件：head <= tail，要包括等号
    2. mid取值：如果head和tail较大，可能会导致溢出，此时应为 mid = head + (tail - head)//2
       或 mid = head + ((tail - head) >> 1)
    3. head和tail的更新: head = mid + 1, tail = mid - 1
'''
def binarySearch_norecurse(arr, obj):
    '''
    简单非递归二分法查找
    '''
    
    head = 0
    tail = len(arr) - 1
    while head <= tail:
        mid = (head + tail)//2
        if arr[mid] == obj:
            return mid
        elif arr[mid] < obj:
            head = mid + 1
        else:
            tail = mid - 1
    return "target don't exist!"

def binarySearch_recurse(arr, head, tail, obj):
    '''
    简单递归二分法查找
    '''
    
    mid = (head + tail)//2
    if head > tail:
        return "target don't exist!"
    if arr[mid] == obj:
        return mid
    elif arr[mid] < obj:
        return binarySearch_recurse(arr, mid+1, tail, obj)
    else:
        return binarySearch_recurse(arr, head, mid-1, obj)

# test
# binarySearch_norecurse
arr = [1, 3, 5, 7, 9]
print(binarySearch_norecurse(arr, 7))
print(binarySearch_norecurse(arr, 10))

# binarySearch_recurse
arr = [1, 3, 5, 7, 9]
print(binarySearch_recurse(arr, 0, len(arr)-1, 7))
print(binarySearch_recurse(arr, 0, len(arr)-1, 10))
```

## 二分查找变形问题

```python
def binarySearch_extension1(arr, obj):
    '''
    二分法查找变形一:查找第一个值等于给定值的元素
    '''
    
    head = 0
    tail = len(arr) - 1
    while head <= tail:
        mid = (head + tail)//2
        if arr[mid] < obj:
            head = mid + 1
        else:
            tail = mid - 1
            
    if head < len(arr) and arr[head] == obj:
        return head
    else:
        return "target don't exist!"

def binarySearch_extension2(arr, obj):
    '''
    二分法查找变形二:查找最后一个值等于给定值的元素
    '''
    
    head = 0
    tail = len(arr) - 1
    while head <= tail:
        mid = (head + tail)//2
        if arr[mid] <= obj:
            head = mid + 1
        else:
            tail = mid - 1
    if head-1 <= len(arr) and arr[head-1] == obj:
        return head - 1
    else:
        return "target don't exist!"

def binarySearch_extension3(arr, obj):
    '''
    二分法查找变形三:查找第一个大于等于给定值的元素
    '''
    
    head = 0
    tail = len(arr) - 1
    while head <= tail:
        mid = (head + tail)//2
        if arr[mid] < obj:
            head = mid + 1
        else:
            if mid == 0 or arr[mid-1] < obj:
                return mid
            else:
                tail = mid - 1
    return 'All the value of element involed in array are small than obj!'

def binarySearch_extension4(arr, obj):
    '''
    二分法查找变形四:查找最后一个小于等于给定值的元素
    '''
    head = 0
    tail = len(arr) - 1
    while head <= tail:
        mid = (head + tail)//2
        if arr[mid] > obj:
            tail = mid - 1
        else:
            if mid == len(arr) - 1 or arr[mid+1] > obj:
                return mid
            else:
                head = mid + 1
    return 'All the value of element involed in array are great than obj!'

# test
arr = [0, 1, 1, 1, 2, 2, 2, 2, 2, 3, 3, 5, 5, 9, 10, 10, 10, 13]
obj = 2

print('-'*10, 'binarySearch_extension1(arr, obj)', '-'*10)
print(binarySearch_extension1(arr, obj))

print('-'*10, 'binarySearch_extension2(arr, obj)', '-'*10)
print(binarySearch_extension2(arr, obj))

print('-'*10, 'binarySearch_extension3(arr, obj)', '-'*10)
print(binarySearch_extension3(arr, obj))

print('-'*10, 'binarySearch_extension3(arr, obj)', '-'*10)
print(binarySearch_extension4(arr, obj))
```

## 二分查找的应用

```python
# 查找轮转数组的轮转点
def findCyclePoint(arr):
    low = 0
    up = len(arr) - 1
    while low < up:
        mid = (low + up)//2
        if arr[mid] > arr[up]:
            low = mid + 1
        elif arr[mid] == arr[up]:
            up = up - 1
        else:
            up = mid
    return arr[low]

# index = (low+k-1)%m

# test
arr = [7,8,9,10,11,121,1,2,3,4,5,6]
print(findCyclePoint(arr))
```

## 二分查找的缺陷

1. 二分查找法的O(log n)让它成为十分高效的算法。不过它的缺陷却也是那么明显的。就在它的限定之上：必须有序，我们很难保证我们的数组都是有序的。
2. 当然可以在构建数组的时候进行排序，可是又落到了第二个瓶颈上：它必须是数组。数组读取效率是O(1)，可是它的插入和删除某个元素的效率却是O(n)。因而导致构建有序数组变成低效的事情。
3. 解决这些缺陷问题更好的方法应该是使用二叉查找树了，最好自然是自平衡二叉查找树了，自能高效的（O(n log n)）构建有序元素集合，又能如同二分查找法一样快速（O(log n)）的搜寻目标数。

# Python实现二叉树及其遍历

```python
# -*- coding: utf-8 -*-

class Node():
    '''
    结点类
    '''
    def __init__(self, value=None, lft=None, rgt = None):
        self.value = value
        self.lft = lft
        self.rgt = rgt

    def __str__(self):
        return str(self.value)

class Tree():
    '''
    树类
    '''
    def __init__(self):
        self.root = Node()
        self.myqueue = [] # 用来存放子树的三个节点，依次为root, left, right

    def add(self, elem):
        node = Node(elem)
        if self.root.value == None: # 树为空，进行赋值
            self.root = node
            self.myqueue.append(self.root)
        else: # 此树不为空，但子节点还没补齐
            treeNode = self.myqueue[0]
            if treeNode.lft == None:
                treeNode.lft = node
                self.myqueue.append(treeNode.lft)
            else:
                treeNode.rgt = node
                self.myqueue.append(treeNode.rgt)
                self.myqueue.pop(0) # 该节点如果存在右子树，将此结点丢弃
    
    # 递归前中后序遍历
    def preOrder(self, root):
        '''
        前序遍历
        '''
        if root == None:
            return
        else:
            print(root.value)
            self.preOrder(root.lft)
            self.preOrder(root.rgt)
    
    def inOrder(self, root):
        '''
        中序遍历
        '''
        if root == None:
            return
        self.inOrder(root.lft)
        print(root.value)
        self.inOrder(root.rgt)

    def postOrder(self, root):
        '''
        后序遍历
        '''
        if root == None:
            return
        self.postOrder(root.lft)
        self.postOrder(root.rgt)
        print(root.value)

    # 非递归前中后序遍历，利用栈实现
    def preOrder_stack(self, root):
        '''
        利用栈实现前序遍历
        '''
        if root == None:
            return
        
        node = root
        mystack = [] # 保存遍历的节点
        while node or mystack:
            while node: # 从根节点开始，先遍历左子树
                print(node.value) # 先打印再入栈
                mystack.append(node)
                node = node.lft
            node = mystack.pop() # 弹出栈顶元素，开始遍历右子树
            node = node.rgt # 每弹出一个结点，紧接着就访问其右子树

    def inOrder_stack(self, root):
        '''
        利用栈实现中序遍历
        '''
        if root == None:
            return

        node = root
        mystack = []
        while node or mystack:
            while node: # 从根节点开始，一直找到左子树到最后一个节点
                mystack.append(node) # 先入栈后打印
                node = node.lft
            node = mystack.pop() # pop顺序: left, root, right
            print(node.value)
            node = node.rgt

    def postOrder_stack(self, root):
        '''
        利用栈实现后序遍历
        '''
        if root == None:
            return
        
        node = root
        mystack1 = [] # 存放正在访问的节点
        mystack2 = [] # 存放后序遍历的逆序
        mystack1.append(node)
        while mystack1:
            node = mystack1.pop()
            if node.lft:
                mystack1.append(node.lft)
            if node.rgt:
                mystack1.append(node.rgt)
            mystack2.append(node)
        
        while mystack2:
            print(mystack2.pop().value)

    def levelOrder(self, root):
        '''
        层次遍历
        '''
        if root == None:
            return 
        myqueue = []
        node = root
        myqueue.append(node)
        while myqueue:
            node = myqueue.pop(0)
            print(node.value)
            if node.lft != None:
                myqueue.append(node.lft)
            if node.rgt != None:
                myqueue.append(node.rgt)

# test
if __name__ == '__main__':
    values = range(10)
    tree = Tree()
    for val in values:
        tree.add(val)
    
    print('-'*10 + '前序遍历' + '-'*10)
    tree.preOrder(tree.root)

    print('-'*10 + '中序遍历' + '-'*10)
    tree.inOrder(tree.root)

    print('-'*10 + '后序遍历' + '-'*10)
    tree.postOrder(tree.root)

    print('-'*10 + '利用stack实现前序遍历' + '-'*10)
    tree.preOrder_stack(tree.root)

    print('-'*10 + '利用stack实现中序遍历' + '-'*10)
    tree.inOrder_stack(tree.root)

    print('-'*10 + '利用stack实现后序遍历' + '-'*10)
    tree.postOrder_stack(tree.root)

    print('-'*10 + 'BFS遍历' + '-'*10)
    tree.levelOrder(tree.root)
```

## 查找操作

1. 取根结点，如果等于要查找的值，返回
2. 如果根结点的值大于要查找的值，那就在左子树中递归查找
3. 如果根结点的值小于要查找的值，那就在右子树中递归查找

```python
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
        
class BinarySearchTree:
    def __init__(self, dataList):
        self.root = Node(dataList[0])
        for val in dataList[1:]:
            self.insert(val)

    def search(self, target):
        node = self.root
        while node and node.val != target:
            if node.val > target:
                node = node.left
            else:
                node = node.right
        return node # while循环终止条件： node == None 或 node.val == target
    
```

## 插入操作

```python
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
        
class BinarySearchTree:
    def __init__(self, dataList):
        self.root = Node(dataList[0])
        for val in dataList[1:]:
            self.insert(val)
            
    def insert(self, value):
        if self.root == None:
            self.root = Node(value)
            return
        pre = None
        node = self.root
        while node:
            pre = node
            if node.val > value:
                node = node.left
            else:
                node = node.right
        new_node = Node(value)
        if pre.val > value:
            pre.left = new_node
        else:
            pre.right = new_node
```

## 删除操作

<img src="http://github.com/lupanlpb/data_stru_and_algo/raw/master/images/二叉查找树删除操作.png" width="500" />

```python
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
        
class BinarySearchTree:
    def __init__(self, dataList):
        self.root = Node(dataList[0])
        for val in dataList[1:]:
            self.insert(val)
            
    def delete(self, value):
        pre = None
        node = self.root
        while node and node.val != value:
            pre = node
            if node.val > value:
                node = node.left
            else:
                node = node.right
        if node == None: return
        
        # 要删除的结点有两个子节点
        if node.left and node.right: # 查找右子树的最小节点
            minP = node.right
            minPP = node # minPP 表示 minP 的父节点
            while minP.left:
                minPP, minP = minP, minP.left
            node.val = minP.val # 将 minP 中的数据替换到 node 中
            pre, node = minPP, minP
            
        # 要删除的结点是叶子节点，或仅有一个叶子结点
        child = node.left if node.left else node.right
        if pre == None:
            self.root = child
        elif pre.left == node:
            pre.left = child
        else:
            pre.right = child
```
